```python
tcr = pandas.read_csv('jack-data/bcc_tcr.txt', sep='\t')
tcr
```

```python
tcr = tcr.drop('cdr3s_aa', axis=1)
tcr
```

```python
tcr_unique = pandas.DataFrame({col: tcr[col].value_counts() for col in tcr.columns}).fillna(0).astype(int)
tcr_unique.head()
```

```python
pre_counts = tcr.loc[tcr.index.str.contains('pre')].apply(pandas.Series.value_counts).fillna(0).sum(axis=1).astype(int)
post_counts = tcr.loc[tcr.index.str.contains('post')].apply(pandas.Series.value_counts).fillna(0).sum(axis=1).astype(int)

# Align the 'pre' and 'post' counts with the index of tcr_unique
tcr_unique['pre'] = tcr_unique.index.map(pre_counts).fillna(0).astype(int)
tcr_unique['post'] = tcr_unique.index.map(post_counts).fillna(0).astype(int)
tcr_unique['pre/post'] = tcr_unique['pre'] / tcr_unique['post']
tcr_unique['pre/post'] = tcr_unique['pre/post'].replace([float('inf'), -float('inf')], pandas.NA)
cols = ['pre', 'post'] + [col for col in tcr_unique.columns if col not in ['pre', 'post']]
tcr_unique = tcr_unique[cols]
```
